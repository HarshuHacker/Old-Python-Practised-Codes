```python
import sys
def minchocolates(li,curr,minimum,minlist):
    currmin=0
    lis=[]
    for i in li:
        a=i%curr
        if(a==0):
            rem=0
        else:            
            b=curr*((i//curr)+1)
            rem=b-i
        lis.append(rem)
        currmin+=rem
    if(currmin<minimum):
        minimum=currmin
        minlist=lis
    return minimum,minlist
li=[int(x) for x in input().split()]
maxi=-sys.maxsize
for i in li:
    maxi=max(maxi,i)
minimum=sys.maxsize
minlist=[]
for curr in range(2,maxi):
    minimum,minlist=minchocolates(li,curr,minimum,minlist)
print(minimum,minlist)
```
